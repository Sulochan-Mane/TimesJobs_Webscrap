# Webscrap TimesJobs

Link : https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=data+analyst&txtLocation=%27

##### Objective :

- Objective is to scrape the Data from the timesjobs for "Data Analytics" and store it in tabular format, so that it can be used in Analytical Projects

## Overview:

### When searched Data Analyst in TimesJobs following window pops up :



### Here there are various whiteboxes which show brief information



#### We can get Job Title, Company name, Experience and Job Location




### Job titles refer to a hyperlink can be accessed which leads to another page




#### From here we can get Job description and skillsets in details





## Libraries Used

- Requests to access html
- BeautifulSoup to ease up the process of scraping
- Pandas to convert the scraped data in Dataframe

## Process

### Extracting the white boxes

- Whiteboxes are under the tag 'li' and class 'clearfix job-bx wht-shd-bx'
- We used Find_all to get a Nested List of White boxes in each page

    [pages[white box1, white box2,.....]]
    
    
- We converted it to 1D list to simplify the code later

### Scraping the Job Titles and Company names

- Job Titles are under the tag h2 and Company names are under h3
- We enumerated the list to get the indices for the jobs

### Scraping Experience, Salary, Location

- Some white boxes had salary and some did not.
- We grouped the white boxes based on that in the lists Normal and Salaried
#### Experience
- First element after finding 'li' tag for each whitebox
- Cleanup is done directly while scraping the data
#### Salary:
- Salaries are provided in second position after finding all the 'li' tag for each whitebox
- Returned empty strings, if salary is not present
#### Location
- For Normal white boxes Location is in the second position after finding all the 'li' tag
- For others, it's 3rd position.

### Scraping Skillset and Job Descriptions

- To access the complete skillset and JDs, we need to accss the hyperlink refered by Job Title
- Created a list called job_links that collects all the webpages refered by the title for each whitebox
#### Skillset
- Skillsets are located under the class 'jd-skill-tag'
- To show all the skillset, we extracted in the form of List and cleaned up using lambda Functions
- List is returned as a string to convert it into dataframe without errors
#### Job Description
- Job Descriptions are under the class 'jd-desc job-description-main'
- We did not cleanup the Job Description, it will be returned as it is

### Creating a Dataset:

- We converted each list into a dataframes, where indices are as generated for each list
- We joined the dataframes on index and we got the final list
